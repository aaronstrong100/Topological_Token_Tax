```python
import tiktoken
from transformers import AutoTokenizer
from huggingface_hub import login
import pandas as pd
from tqdm import tqdm

test_string = "معالجة اللغات الطبيعية (NLP) is difficult for agglutinative words like Muvaffakiyetsizleştiricileştiriveremeyebileceklerimizdenmişsinizcesine."


llama2_tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-hf")
llama3_tokenizer = AutoTokenizer.from_pretrained("meta-llama/Meta-Llama-3-8B")
aya_tokenizer = AutoTokenizer.from_pretrained("CohereForAI/aya-23-8B")
gpt4o_tokenizer = tiktoken.get_encoding("o200k_base")
gemma_tokenizer = AutoTokenizer.from_pretrained("google/gemma-3-4b-pt")

def llama2_tokenize(text):
    llama2_tokens = llama2_tokenizer.tokenize(text)
    return len(llama2_tokens)

def llama3_tokenize(text):
    llama3_encoded = llama3_tokenizer.encode(text)
    llama3_tokens = [llama3_tokenizer.decode([t]) for t in llama3_encoded]
    return len(llama3_tokens)

def aya_tokenize(text):
    aya_tokens = aya_tokenizer.tokenize(text)
    return len(aya_tokens)

def gpt4o_tokenize(text):
    gpt4o_encoded = gpt4o_tokenizer.encode(text)
    gpt4o_tokens = [gpt4o_tokenizer.decode([t]) for t in gpt4o_encoded]
    return len(gpt4o_tokens)

def gemma_tokenize(text):
    gemma_tokens = gemma_tokenizer.encode(text)
    return len(gemma_tokens)

def tokenize_languages():
    tokenizers = {
        "LLaMA 2": llama2_tokenize,
        "LLaMA 3": llama3_tokenize,
        "Aya": aya_tokenize,
        "GPT-4o": gpt4o_tokenize,
        "Gemma-2b": gemma_tokenize
    }

    flores_languages = pd.read_csv("flores_languages.csv")

    pbar = tqdm(range(len(flores_languages)), total=len(flores_languages))

    for i in pbar:
        iso_code = flores_languages.loc[i, "ISO_Code"]
        language_name = flores_languages.loc[flores_languages["ISO_Code"] == iso_code, "Language_Name"].values[0]
        pbar.set_description("Processing " + language_name + " (" + iso_code + ")")
        text_file_name = f"flores_datasets/{iso_code}.txt"
        text = ""
        with open(text_file_name, "r", encoding="utf-8") as f:
            text = f.read()
        for tokenizer_name, tokenizer_pipeline in tokenizers.items():
            flores_languages.loc[flores_languages["ISO_Code"] == iso_code, tokenizer_name + "_Token_Count"] = tokenizer_pipeline(text)
    flores_languages.to_csv("flores_token_counts.csv", index=False)

tokenize_languages()
```

```python
def count_sentences(text):
    return len(text.split('.'))

def language_sentence_counts():
    flores_languages = pd.read_csv("flores_token_counts.csv")
    iso_codes = list(flores_languages["ISO_Code"])

    for iso_code in iso_codes:
        text_file_name = f"flores_datasets/{iso_code}.txt"
        sentences = 0
        with open(text_file_name, "r", encoding="utf-8") as f:
             for line in f:
                 sentences += 1
        flores_languages.loc[flores_languages["ISO_Code"] == iso_code, "sentences"] = sentences
        print(f"{iso_code} has {sentences} sentences.")
    flores_languages.to_csv("flores_token_counts.csv", index=False)

def normalize_tokenizer_counts():
    flores_languages = pd.read_csv("flores_token_counts.csv")
    token_columns = [col for col in flores_languages.columns if col.endswith("_Token_Count")]
    for col in token_columns:
        flores_languages[col + "_Per_Sentence"] = flores_languages[col] / flores_languages["sentences"]
    flores_languages.to_csv("flores_token_counts.csv", index=False)

language_sentence_counts()
normalize_tokenizer_counts()
```
