```python
import os
import urllib.request
import tarfile
import pandas as pd
import shutil

# 1. Ensure the output directory exists
os.makedirs("flores_datasets", exist_ok=True)

archive_url = "https://tinyurl.com/flores200dataset"
archive_path = "flores200_dataset.tar.gz"

print("Checking FLORES-200 archive...")
if not os.path.exists(archive_path):
    print("Downloading the official FLORES-200 archive (this may take a minute)...\n")
    req = urllib.request.Request(
        archive_url, 
        headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
    )
    try:
        with urllib.request.urlopen(req) as response, open(archive_path, 'wb') as out_file:
            shutil.copyfileobj(response, out_file)
        print(" ✓ Archive downloaded successfully.\n")
    except Exception as e:
        print(f" ✗ Critical Download Failure: {e}")
else:
    print(" ✓ Archive already exists locally.\n")

print("Mapping archive contents dynamically...\n")

try:
    with tarfile.open(archive_path, "r:gz") as tar:
        # Build a dictionary of ALL .devtest files inside the tarball.
        # os.path.basename strips away any messy internal folder structures,
        # mapping exactly 'ace_Arab.devtest' directly to its file object.
        devtest_members = {
            os.path.basename(m.name): m 
            for m in tar.getmembers() 
            if m.name.endswith(".devtest")
        }

        print("Extracting requested languages...\n")
        
        for i in range(len(flores_df)):
            # .strip() handles any hidden whitespaces in your DataFrame strings
            flores_code = str(flores_df.loc[i, "Language_Code"]).strip()
            iso_code = str(flores_df.loc[i, "ISO_Code"]).strip()
            
            output_filepath = os.path.join("flores_datasets", f"{iso_code}.txt")
            print(f"Extracting {flores_code} -> Saving as {iso_code}.txt")
            
            target_filename = f"{flores_code}.devtest"
            
            # Check the dictionary instead of guessing the path
            if target_filename in devtest_members:
                member = devtest_members[target_filename]
                f_in = tar.extractfile(member)
                
                if f_in is not None:
                    # Decode the bytes to a standard UTF-8 string
                    content = f_in.read().decode("utf-8")
                    
                    with open(output_filepath, "w", encoding="utf-8") as f_out:
                        f_out.write(content)
                    print("  ✓ Saved successfully.")
                else:
                    print(f"  ✗ Failed to read {target_filename} from archive.")
            else:
                print(f"  ✗ Failed: {flores_code} not found in the archive.")
                
    print("\nBatch extraction complete!")
    
except FileNotFoundError:
    print("\nExtraction aborted: The archive file was not found.")
except Exception as e:
    print(f"\nCritical Error: {e}")
```
